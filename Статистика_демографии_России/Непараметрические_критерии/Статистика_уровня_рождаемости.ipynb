```python
# Проверка гипотезы о равенстве дисперсий для коэффициента  (уровень рождаемости) в следующие периоды: (1990, 2000), (2001, 2010), (2011, 2017), используя непараметрические критерии по Москве и Московской области
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
import seaborn as sns
df=pd.read_csv("russian_demography.csv")
df1=df[(df['region'].isin(['Moscow','Moscow Oblast'])) & ((df['year']>=1990) & (df['year']<=2000))]['birth_rate']
df2=df[(df['region'].isin(['Moscow','Moscow Oblast'])) & ((df['year']>=2001) & (df['year']<=2010))]['birth_rate']
df3=df[(df['region'].isin(['Moscow','Moscow Oblast'])) & ((df['year']>=2011) & (df['year']<=2017))]['birth_rate']
# Визуализация
fig,ax=plt.subplots(1,3,figsize=(15,5))
ax[0].hist(df1)
mean1 = df1.mean()
ax[0].axvline(df1.mean(), color='red', linewidth=2)
ax[1].hist(df2)
ax[1].axvline(df2.mean(), color='red', linewidth=2)
ax[2].hist(df3)
ax[2].axvline(df3.mean(), color='red', linewidth=2)
ax[0].set_title('1990-2000')
ax[1].set_title('2001-2010')
ax[2].set_title('2011-2017')
ax[0].set_xlabel('birth_rate')
ax[1].set_xlabel('birth_rate')
ax[2].set_xlabel('birth_rate')
fig.tight_layout(pad=0.5)
# Непараметрические критерии для дисперсии
# Критерий Левена
print("Критерий Левена:", scipy.stats.levene(df1,df2,df3))
# Гипотезу о равенстве дисперсий не отвергаем p=0.48 > a=0.05
# Критерий Флигнера-Киллина
print("Критерий Флигнера-Киллина:",scipy.stats.fligner(df1,df2,df3))
# Гипотезу о равенстве дисперсий не отвергаем p=0.45 > a=0.05
```
