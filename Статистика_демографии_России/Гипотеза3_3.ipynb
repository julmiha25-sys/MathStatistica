```python
import pandas as pd
from scipy.stats import *
import scipy
from scipy.stats import f_oneway
df=pd.read_csv('russian_demography.csv', header=0)
l=['Moscow','Moscow Oblast']
df1=df[df['region'].isin(l)].copy()
df1=df1.dropna(subset=['urbanization', 'region', 'year'])
group_2000_2005 = df1[(df1['year'] >= 2000) & (df1['year'] <= 2005)]['urbanization'].values
group_2006_2011 = df1[(df1['year'] >= 2006) & (df1['year'] <= 2011)]['urbanization'].values
group_2012_2017 = df1[(df1['year'] >= 2012) & (df1['year'] <= 2017)]['urbanization'].values
groups = [group_2000_2005, group_2006_2011, group_2012_2017]
group_names = ['2000-2005', '2006-2011', '2012-2017']
# Проверка равенство дисперсий по критерию Бартлетта
bartlett(*groups)
# Дисперсии равны - односторонний дисперсионный анализ
bartlett(*groups),scipy.stats.f_oneway(*groups, axis=0)
```
