```python
import pandas as pd
from scipy.stats import *
import scipy
df=pd.read_csv('russian_demography.csv', header=0)
l=['Amur Oblast','Altai Krai','Astrakhan Oblast']
df1=df[df['region'].isin(l)]
df2=df1[['region','urbanization']]
df2=df2.dropna(subset=['urbanization', 'region'])
groups=[group['urbanization'].values for name, group in df2.groupby('region')]
# Проверка равенство дисперсий по критерию Бартлетта
bartlett(*groups)
# Дисперсии не равны - используем критерий Говерна
scipy.stats.alexandergovern(*groups, nan_policy='propagate')
# Тест Тьюкки для средних
bartlett(*groups),scipy.stats.alexandergovern(*groups, nan_policy='propagate') ,tukey_hsd(*groups).pvalue
```
