```python
import pandas as pd
from scipy.stats import *
import scipy
import numpy as np
from scipy.stats import binomtest

df1=pd.read_csv('Salary.csv', header=0)
# Удаление пропусков
#df1=df.dropna()
# Преобразование типа
df1.loc[:, 'Опыт работы']=df1['Опыт работы'].astype(int)
# Количество записей - размер выборки
n_v=len(df1.dropna(subset=['Уровень образования']))
alpha=0.01
d=0.15
# Определение доли людей со средним профессиональным образованием
df_v=df1[df1['Уровень образования'].isin(['среднее профессиональное'])]
result = binomtest(k=len(df_v), n=n_v, p=0.15, alternative='two-sided')
r=result.proportion_ci(confidence_level=1-alpha)
print("Доверительный интервал от", r[0], " до ", r[1])
print("Выборочная доля:",len(df_v)/len(df1))
if d>r[0] and d<r[1]:
  print(f"Гипотетическая доля {d} входит в доверительный интервал - Гипотезу Не ОТВЕРГАЕМ")
else:
  print(f"Гипотетическая доля {d} не входит в доверительный интервал - Гипотезу ОТВЕРГАЕМ")
```
