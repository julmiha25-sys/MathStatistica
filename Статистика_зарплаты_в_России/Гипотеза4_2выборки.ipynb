```python
import pandas as pd
import scipy
import numpy as np
from scipy.stats import *
from math import *
def opis_stat(df):
    # Список всех параметров описательной ститистики
    q=scipy.stats.describe(df, axis=0, ddof=1, bias=True)
    print(f"Среднее: {q.mean: .2f}")
    print(f"Несм. дисперсия: {q.variance: .2f}")
    print(f"Количество: {q.nobs: .2f}")
    return (q.mean,q.variance,q.nobs)

df1=pd.read_csv('Salary.csv', header=0)
# Преобразование типа
df1.loc[:, 'Опыт работы']=df1['Опыт работы'].astype(int)
# Датафреймы по возрасту и зарплате
df_m=df1[df1['Возраст']<25]['Зарплата']
df_f=df1[df1['Возраст']>55]['Зарплата']
print("Статистика зарплат с возрастом до 25 лет:")
mean_m, var_m, m = opis_stat(df_m)
print("\nСтатистика зарплат с возрастом свыше 55 лет:")
mean_f, var_f, fe = opis_stat(df_f)
# Критерий Фишера
F = var_m/var_f
print("\nКритерий Фишера", F)
# Критическая область
alpha=0.05
k=f.ppf(1-alpha, m-1, fe-1)
print(f"Критическая область: ({k}; +∞)")
if F < k:
    print("Критерий Фишера не лежит в критической области - нет оснований отвергнуть гипотезу о равенстве дисперсий")
else:
    print("Критерий Фишера лежит в критической области - есть основания отвергнуть гипотезу о равенстве дисперсий")
# Стандартная ошибка
s_p=((m-1)*var_m+(fe-1)*var_f)/(m+fe-2)
delta=t.ppf(1-alpha/2, m+fe-2)*(sqrt(s_p)*sqrt(1/fe+1/m))
a=(mean_m-mean_f)-delta
b=(mean_m-mean_f)+delta
# Доверительный интервал
print(f'\nДоверительный интервал: ({a:.2f};{b:.2f})')
print(f"Разница средних зарплат для возрастных групп до 25 лет и свыше 55 лет: {(mean_m-mean_f):.2f}")
if 0<(mean_m-mean_f)-delta or 0>(mean_m-mean_f)+delta:
  print('Есть основания отвергнуть гипотезу о равенстве средних')
else:
  print('Нет оснований отвергнуть гипотезу о равенстве средних')
# Проверка вероятности
T=(mean_m-mean_f)/(sqrt(s_p)*sqrt(1/fe+1/m))
p_value = t.sf(T, m+fe-2)
print(f"Вероятность гипотезы p-value: {p_value:.4f}")
if p_value < alpha:
    print("Есть основания отвергнуть гипотезу о равенстве средних")
else:
    print("Нет оснований отвергнуть гипотезу о равенстве средних")
```
