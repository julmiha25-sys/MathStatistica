```python
import pandas as pd
from scipy.stats import norm
import scipy
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('Salary.csv', header=0)
# Преобразование типа
df.loc[:, 'Опыт работы']=df['Опыт работы'].astype(int)
#df0=df.dropna(subset=['Уровень образования'])
df1=df.sample(n=int(len(df)/2))
df2=df.drop(index=list(df1.index))
c1=len(df1)
m1=len(df1[df1['Уровень образования'] == 'среднее профессиональное'])
d1=m1/c1
c2=len(df2)
m2=len(df2[df2['Уровень образования'] == 'среднее профессиональное'])
d2=m2/c2
print(f"Доли людей со среднепрофессиональным образованием: {d1:.3f}, {d2:.3f}")
alpha=1-0.98
# Проверка репрезентативности выборок
if c1*d1>=5 and c1*(1-d1)>=5 and c2*d2>=5 and c2*(1-d2)>=5:
  print("Выборка репрезентативна")
else:
  print("Выборка НЕ репрезентативна")
# Статистика z-теста
p=(m1+m2)/(c1 + c2)
z=(d1-d2)/((p*(1-p)/c1+p*(1-p)/c2)**0.5)
print("Статистика z-теста",z)
p_value=2*norm.sf(abs(z))
print(f"Вероятность гипотезы p-value: {p_value:.4f}")
# Вывод гипотезы
if p_value < alpha:
    print("Отвергаем H0: доли людей со средним профессиональным образованием в этих двух выборках НЕ равны")
else:
    print("Не отвергаем H0: доли людей со средним профессиональным образованием в этих двух выборках равны")
```
