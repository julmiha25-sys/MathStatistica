```python
import pandas as pd
import scipy
import numpy as np
from scipy.stats import *
from math import *

def opis_stat(df):
    # Список всех параметров описательной ститистики
    q=scipy.stats.describe(df, axis=0, ddof=1, bias=True)
    print(f"Среднее: {q.mean: .2f}")
    print(f"Несм. дисперсия: {q.variance: .2f}")
    print(f"Количество: {q.nobs: .2f}")
    return (q.mean,q.variance,q.nobs)

df1=pd.read_csv('Salary.csv', header=0)
# Преобразование типа
df1.loc[:, 'Опыт работы']=df1['Опыт работы'].astype(int)
# Датафреймы по полу
df_m=df1[df1['Пол']=='мужской']['Зарплата']
df_f=df1[df1['Пол']=='женский']['Зарплата']
print("Статистика по мужчинам:")
mean_m, var_m, m = opis_stat(df_m)
print("\nСтатистика по женщинам:")
mean_f, var_f, fe = opis_stat(df_f)
# Критерий Фишера
F=var_m/var_f
print("\nКритерий Фишера", F)
# Критическая область
alpha=0.01
k=f.ppf(1-alpha,m,fe)
print(f"Критическая область: ({k};+∞)")
if F<k:
  print("Критерий Фишера не лежит в критической области - нет оснований отвергнуть гипотезу о равенстве дисперсий")
else:
  print("Критерий Фишера лежит в критической области - есть основания отвергнуть гипотезу о равенстве дисперсий")
# Вероятность того, что статистика примет значение большее или равное наблюдаемому
p_value=f.sf(F,m-1,fe-1)
print(f"Вероятность гипотезы p-value: {p_value:.4f}")
# Вывод гипотезы
if p_value < alpha:
  print("Есть основания отвергнуть гипотезу о равенстве дисперсий")
else:
  print("Нет оснований отвергнуть гипотезу о равенстве дисперсий")
# Если дисперсии равны, то используем критерий Стьюдента для средних
sp=((m-1)*var_m + (fe-1)*var_f)/(m+fe-2)
T=(mean_m-mean_f)/(sqrt(sp)*sqrt(1/m+1/fe))
print(f"Критерий Стьюдента {T:.4f}")
# Критическая область распределения Стюдента
k=t.ppf(1-alpha,m+fe-2)
print(f"Критическая область: ({k};+∞)")
if T<k:
  print("Критерий Стьюдента не лежит в критической области - нет оснований отвергнуть гипотезу о равенстве средних зарплат мужчины и женщины")
else:
  print("Критерий Стьюдента лежит в критической области - есть основания отвергнуть гипотезу о равенстве средних зарплат мужчины и женщины")
# Вероятность того, что статистика примет значение большее или равное наблюдаемому
p_value=t.sf(T,m+fe-2)
print(f"Вероятность гипотезы p-value: {p_value:.4f}")
# Вывод гипотезы
if p_value < alpha:
  print("Есть основания отвергнуть гипотезу о равенстве средних зарплат мужчины и женщины")
else:
  print("Нет оснований отвергнуть гипотезу о равенстве средних зарплат мужчины и женщины")
```
