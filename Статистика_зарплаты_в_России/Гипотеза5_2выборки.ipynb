```python
import pandas as pd
import scipy
import numpy as np
from scipy.stats import ttest_ind
from scipy.stats import norm
df1=pd.read_csv('Salary.csv', header=0)
# Преобразование типа
df1.loc[:, 'Опыт работы']=df1['Опыт работы'].astype(int)
# Датафреймы по зарплате-городу
col_msk=len(df1[df1['Регион']=='Москва'])
col_pit=len(df1[df1['Регион']=='Санкт-Петербург'])
mean_msk=df1[(df1['Регион']=='Москва')]['Зарплата'].mean()
mean_pit=df1[(df1['Регион']=='Санкт-Петербург')]['Зарплата'].mean()
z_msk=len(df1[(df1['Регион']=='Москва') & (df1['Зарплата']> mean_msk)]['Зарплата'])
z_pit=len(df1[(df1['Регион']=='Санкт-Петербург') & (df1['Зарплата']> mean_pit)]['Зарплата'])
d_msk=z_msk/col_msk
d_pit=z_pit/col_pit
print(f"Доли москвичей и питерцев с зарплатой выше средней: {d_msk:.2f}, {d_pit:.2f}")
# Проверка репрезентативности выборок
if col_msk*d_msk>=5 and col_msk*(1-d_msk)>=5 and col_pit*d_pit>=5 and col_pit*(1-d_pit):
  print("Выборки репрезентативны")
else:
  print("Выборки  НЕ репрезентативны")
# Статистика z-теста
p=(z_msk+z_pit)/(col_msk+col_pit)
z=(d_msk-d_pit)/((p*(1-p)/col_msk+p*(1-p)/col_pit)**0.5)
print("Статистика z-теста",z)
p_value=norm.sf(z)
print(f"Вероятность гипотезы p-value: {p_value:.4f}")
# Вывод гипотезы
alpha=0.1
if p_value < alpha:
    print("Не отвергаем H0: ЕСТЬ различия долей людей с высокой зарплатой (выше среднего) в выборке между Москвой и Санкт-Петербургом")
else:
    print("Отвергаем H0: НЕТ различия в долях людей с высокой зарплатой (выше среднего) в выборке между Москвой и Санкт-Петербургом")
```
