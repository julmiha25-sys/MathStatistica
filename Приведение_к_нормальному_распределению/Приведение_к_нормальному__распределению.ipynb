```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
Salary = pd.read_csv('Salary.csv', header=0, sep=',')
# Логарифмическое преобразование
log_transform = np.log(Salary['Зарплата'])
# Десятичный логарифм от данных
log10_transform = np.log10(Salary['Зарплата'])
# Преобразование квадратного корня
sqrt_transform = np.sqrt(Salary['Зарплата'])
# Преобразование кубического корня
cbrt_transform = np.cbrt(Salary['Зарплата'])
# Преобразование обратная дробь
invers_transform = 1/Salary['Зарплата']
# Визуализация
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes[0, 0].set_title(f'Исходные данные, коэф.ассим. {Salary['Зарплата'].skew():.2f} ')
Salary['Зарплата'].hist(bins=10, ax=axes[0, 0])
log_transform.hist(bins=10, ax=axes[0, 1])
axes[0, 1].set_title(f'log_transform, коэф.ассим. {log_transform.skew():.2f}')
log10_transform.hist(bins=10, ax=axes[0, 2])
axes[0, 2].set_title(f'log10_transform, коэф.ассим. {log10_transform.skew():.2f}')
sqrt_transform.hist(bins=10, ax=axes[1, 0])
axes[1, 0].set_title(f'sqrt_transform, коэф.ассим. {sqrt_transform.skew():.2f}')
cbrt_transform.hist(bins=10, ax=axes[1, 1])
axes[1, 1].set_title(f'cbrt_transform, коэф.ассим. {cbrt_transform.skew():.2f}')
invers_transform.hist(bins=10, ax=axes[1, 2])
axes[1, 2].set_title(f'invers_transform, коэф.ассим. {invers_transform.skew():.2f}')
plt.tight_layout()
plt.show()
print('Исходные:', Salary['Зарплата'].skew())
print('log_transform:', log_transform.skew())
print('log10_transform:', log10_transform.skew())
print('sqrt_transform:', sqrt_transform.skew())
print('cbrt_transform:', cbrt_transform.skew())
print('invers_transform:', invers_transform.skew())
```
