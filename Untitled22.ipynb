```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

samples = np.random.normal(2, 8, 1000)
res = stats.relfreq(samples, numbins=25)

x = res.lowerlimit + np.linspace(0, res.binsize*res.frequency.size, res.frequency.size)

fig = plt.figure(figsize=(5, 4))
ax = fig.add_subplot(1, 1, 1)
ax.bar(x, res.frequency, width=res.binsize)
ax.set_title('Relative frequency histogram')
ax.set_xlim([x.min(), x.max()])

plt.show()
```

```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt

# Параметры
a, b = 2, 6
loc, scale = 3, 10
size = 2000

# Генерируем выборку
sample = beta.rvs(a, b, loc=loc, scale=scale, size=size)

# Создаем фигуру
fig, ax = plt.subplots(figsize=(10, 6))

# 1. Строим гистограмму (нормированную)
ax.hist(sample, bins=30, density=True, alpha=0.6,
        color='lightblue', edgecolor='black',
        label='Эмпирическая выборка (гистограмма)')

# 2. Строим теоретическую плотность вероятности
# Создаем много точек для плавной кривой
x = np.linspace(loc, loc + scale, 1000)
# Вычисляем плотность вероятности в этих точках
y = beta.pdf(x, a, b, loc=loc, scale=scale)

ax.plot(x, y, 'r-', linewidth=2, label='Теоретическая плотность (PDF)')

# Настройки графика
ax.set_title(f'Бета-распределение (a={a}, b={b}) со смещением {loc} и разбросом {scale}')
ax.set_xlabel('Значение')
ax.set_ylabel('Плотность вероятности')
ax.legend()
ax.grid(True, alpha=0.3)

plt.show()
```
