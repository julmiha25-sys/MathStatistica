```python
import scipy.stats as stats
from scipy.stats import mannwhitneyu
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import *
df = pd.read_csv("daily-website-visitors.csv")
df['Date'] = pd.to_datetime((df['Date'].str.replace("/","-")), format='%m-%d-%Y')
df['Unique.Visits']=pd.to_numeric(df['Unique.Visits'].str.replace(",",""))
df1=df[pd.to_datetime(df['Date']).dt.year==2016]['Unique.Visits']
df2=df[pd.to_datetime(df['Date']).dt.year==2017]['Unique.Visits']
df3=df[pd.to_datetime(df['Date']).dt.year==2018]['Unique.Visits']
# Визуализация
fig, axs = plt.subplots(1,3,figsize=(10,3))
axs[0].hist(df1)
axs[0].set_title('2016')
axs[0].set_xlabel('Unique.Visits')
axs[1].hist(df2)
axs[1].set_title('2017')
axs[1].set_xlabel('Unique.Visits')
axs[2].hist(df3)
axs[2].set_title('2018')
axs[2].set_xlabel('Unique.Visits')
# Применение теста Андерсона-Дарлинга
print(f"Тест Андерсона-Дарлинга : {anderson_ksamp([df1, df2, df3])}")
# p_value=0.001 < a=0.05 - гипотезу о согласованности выборок 2016/2017/2018 отвергаем
```
